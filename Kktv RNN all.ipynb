```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout, BatchNormalization, LSTM
from keras.callbacks import Callback, EarlyStopping
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_RNN, y_RNN, test_size=0.2)
```

```python
model = Sequential()
model.add(LSTM(4, return_sequences=False, input_shape=(X_RNN.shape[1], X_RNN.shape[2])))
model.add(Dense(28, activation='sigmoid'))
```

```python
model.summary()
```

```python
class PlotLosses(Callback):
    def on_train_begin(self, logs={}):
        self.i = 0
        self.x = []
        self.losses = []
        self.val_losses = []
        
        self.fig = plt.figure()
        
        self.logs = []

    def on_epoch_end(self, epoch, logs={}):
        
        self.logs.append(logs)
        self.x.append(self.i)
        self.losses.append(logs.get('loss'))
        self.val_losses.append(logs.get('val_loss'))
        self.i += 1
        
        clear_output(wait=True)
        plt.plot(self.x, self.losses, label="loss")
        plt.plot(self.x, self.val_losses, label="val_loss")
        plt.legend()
        plt.show();
        
plot_losses = PlotLosses()
```

```python
early_stop = EarlyStopping(monitor='val_loss', min_delta=0, patience=5, verbose=0, mode='auto')
```

```python
model.compile(loss='binary_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
model.fit(X_train, y_train,
          batch_size=128,
          epochs=30,
          verbose=1,
          callbacks=[plot_losses, early_stop],
          validation_data=(X_test, y_test))
#history = model.fit(X, y, epochs=400, batch_size=len(X), verbose=2)
# plot metrics
#pyplot.plot(history.history['acc'])
#pyplot.show()
score = model.evaluate(X_test, y_test, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
